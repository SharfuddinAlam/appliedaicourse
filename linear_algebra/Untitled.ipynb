```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

from PIL import Image


# Experiment-2
bnw_img = Image.open('/Users/krishna/Desktop/venkat_krishna_turlapati/training_material/appliedaicourse/datasets/test.jpg')
imggray = bnw_img.convert('LA')
plt.figure(figsize=(9, 6))
plt.imshow(imggray);

# Now I'll convert the image data into a numpy matrix, plotting the result to show the data is unchanged.
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
plt.figure(figsize=(9,6))
plt.imshow(imgmat, cmap='gray')


U, sigma, V = np.linalg.svd(imgmat)


for i in range(1, 4):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()


# plot U
for i in range(1, 700, 100):
    reconstimg = np.matrix(U[:,:i])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()



# plot V
for i in range(1, 700, 100):
    reconstimg = np.matrix(V[:,:i])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```
